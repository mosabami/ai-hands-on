**Train & Test Splitting**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
X_train
```

**Let's Import Linear Model from **sklearn** and start using it.**

[Read Here in Details](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

Do we have to stanardize/normalize the data ?

* We always prefer data to be normilize to apply gradient descent.
* Standardize means: z = (x - u) / s. Read more about it [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html) in details.

```python
# Standardizing the dataset

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

**After standerdizing the data, I just fit the data and then transform it.**

**!! Remember:** This method combines the `fit()` and `transform()` operations into a single step. Primarily used on the training data to learn the transformation parameters and apply them in one go, optimizing efficiency.

#### What `.fit_transform()` does?

 It first calculates the necessary parameters from the provided data (e.g., mean and standard deviation for scaling, or the imputation value for missing data), and then immediately applies the transformation to the same data using those learned parameters.

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

**`.transform()`** Used on testing data or any new, unseen data to apply the same transformations learned from the training data.

**To take your X_train Data to original shape, you can just inverse it.**

```python
scaler.inverse_transform(X_train)
```

```python
from sklearn.linear_model import LinearRegression
##cross validation
from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train, y_train)
```

**Read Cross Validation Details [here](https://scikit-learn.org/stable/modules/cross_validation.html).**

**Check Model Evaluation Details [here](https://scikit-learn.org/stable/modules/model_evaluation.html)**

```python
MSE = cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
MSE
```

* `cv = 5`: it means the dataset will be split into 5 equal-sized, random partitions, also known as "folds". I will get 5- different type of cross validation accuracy.
* `scoring='neg_mean_squared_error'`:

    **Mean Squared Error (MSE)** is standard metric for regression tasks that measures the average squared difference between the estimated values and the actual values. A lower MSE indicates a better-fitting model, with 0.0 being the best possible value.

    **The Negation:** Scikit-learn's cross-validation and hyperparameter tuning tools are designed to work by maximizing the evaluation result. Since MSE is an error metric (where lower is better), it is multiplied by -1 to convert it into a "score" metric (where higher is better).

```python
cross_val_score(regression, X_train, y_train, scoring='neg_mean_squared_error', cv=10)
```

```python
np.mean(MSE)
```

## Implement Ridge Regression

### Subtask:
Implement Ridge Regression, a regularized linear model, which can be more robust to multicollinearity and potentially improve performance compared to standard Linear Regression, especially with a low R-squared score. We will use `sklearn.linear_model.Ridge`.


**Reasoning**:
The subtask requires importing the `Ridge` class. This command will import `Ridge` from `sklearn.linear_model`.



```python
from sklearn.linear_model import Ridge
print("Ridge class imported successfully.")
```

**Reasoning**:
Now that the `Ridge` class is imported, the next step is to instantiate a `Ridge` object as specified by the subtask.



```python
ridge_regression = Ridge()
print("Ridge regression model instantiated with default parameters.")
```

**Reasoning**:
With the `Ridge` model instantiated, the next step is to train it using the scaled training data. This is consistent with how the `LinearRegression` model was trained earlier.



```python
ridge_regression.fit(X_train, y_train)
print("Ridge regression model trained successfully.")
```

**Reasoning**:
The `NameError` for `X_train` and `y_train` occurred because the `train_test_split` and `StandardScaler` operations were performed before `X` and `y` were correctly defined based on the diabetes dataset's description. To resolve this, I need to re-execute the data splitting and standardization steps with the now correctly assigned `X` (features) and `y` (target) variables. This ensures `X_train`, `X_test`, `y_train`, and `y_test` are properly defined and scaled before training the Ridge model.



```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print("X_train, X_test, y_train, y_test have been re-split and standardized.")
```

**Reasoning**:
Now that `X_train` and `y_train` are correctly defined and scaled, I need to re-train the `ridge_regression` model. This step aligns with the subtask of implementing Ridge Regression.



```python
ridge_regression.fit(X_train, y_train)
print("Ridge regression model re-trained successfully.")
```

**Reasoning**:
After training the `ridge_regression` model, the next step is to evaluate its performance using cross-validation with negative mean squared error, as specified in the main task. This will provide a measure of the model's accuracy on different folds of the training data.



```python
ridge_MSE = cross_val_score(ridge_regression, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
print("Cross-validation MSE for Ridge Regression calculated.")
ridge_MSE
```

**Reasoning**:
After calculating the individual cross-validation MSEs for Ridge Regression, I need to compute their mean to get a single, representative metric of the model's performance, similar to how the Linear Regression model's MSE was summarized.



```python
np.mean(ridge_MSE)
print(f"Mean Cross-validation MSE for Ridge Regression: {np.mean(ridge_MSE)}")
```

**Reasoning**:
To compare the Ridge Regression model's performance with the initial Linear Regression model using the R-squared score, I first need to make predictions on the test set using the trained `ridge_regression` model.



```python
ridge_pred = ridge_regression.predict(X_test)
print("Predictions for Ridge Regression on the test set generated.")
ridge_pred
```

**Reasoning**:
Now that predictions for Ridge Regression are available, the next logical step is to calculate the R-squared score, as specified in the main task, to quantify how well the model's predictions align with the actual `y_test` values.



```python
from sklearn.metrics import r2_score

ridge_score = r2_score(y_test, ridge_pred)
print(f"R-squared score for Ridge Regression: {ridge_score}")
```

**Reasoning**:
The main task asks to visualize the residuals after evaluating the Ridge Regression model. This plot will help in understanding the distribution of the errors.



```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.displot(ridge_pred - y_test, kind = 'kde')
plt.title('Distribution of Residuals for Ridge Regression')
plt.xlabel('Residuals (Predicted - Actual)')
plt.ylabel('Density')
print("Residuals for Ridge Regression visualized with a KDE plot.")
```
